```python
import requests
import pandas as pd
import time

df_list = []

for i in range(1, 50):
    print(f"Fetching page {i}...")

    url = f"https://api.coingecko.com/api/v3/coins/markets" \
          f"?vs_currency=usd&order=market_cap_desc&per_page=100&page={i}"

    response = requests.get(url)
    data = response.json()

    if not isinstance(data, list):
        print("API returned error:", data)
        break

    if len(data) == 0:
        print("No more data. Stopping.")
        break

    
    temp_df = pd.DataFrame(data, columns=[
        'id','name','current_price','market_cap','total_volume',
        'max_supply','price_change_percentage_24h','circulating_supply'
    ])

    df_list.append(temp_df)
    time.sleep(1)  

df = pd.concat(df_list, ignore_index=True)

df.head()

```

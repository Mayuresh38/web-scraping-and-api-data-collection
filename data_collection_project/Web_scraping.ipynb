```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
df_list = []

for j in range(1, 51):
    if j == 1:
        url = 'https://books.toscrape.com/'
    else:
        url = f'https://books.toscrape.com/catalogue/page-{j}.html'

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
                       AppleWebKit/537.36 (KHTML, like Gecko) \
                       Chrome/120.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers=headers)
    response.encoding = 'utf-8'
    webpage = response.text
    
    soup = BeautifulSoup(webpage,"lxml")

    company = soup.find_all('article', class_='product_pod')

    name = []
    rating = []
    price = []
    stock = []

    for i in company:
        #Name
        name_tag = i.find('h3').find('a')
        name.append(name_tag.get('title').strip())

        #Rating
        rating_tag = i.find('p', class_="star-rating")
        rating.append(rating_tag['class'][1])

        #Price
        price_tag = i.find('p', class_='price_color')
        price.append(price_tag.text.strip())

        #Stock
        stock_tag = i.find('p', class_='availability')
        if stock_tag:
            stock.append(stock_tag.text.strip())
        else:
            stock.append("N/A")

    d = {
        "Name": name,
        "Rating": rating,
        "Price": price,
        "Stock": stock
    }

    temp_df = pd.DataFrame(d)
    df_list.append(temp_df)

    print(f"Page {j} done - {len(temp_df)} books scraped")

df = pd.concat(df_list, ignore_index=True)
print(f"\n Total books: {len(df)}")

df.to_csv("Book_dataset.csv", index=False)
    



                            
       
```
